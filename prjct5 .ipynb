```python
# prompt: show  salary trends, this dataset offers a wealth of information to explore
# and analyze.

import matplotlib.pyplot as plt
# Create a line plot of average salary by job role
df.groupby('job_title')['salary_in_usd'].mean().plot(kind='line')
plt.title('Average Salary by Job Role')
plt.xlabel('Job Role')
plt.ylabel('Average Salary (USD)')
plt.xticks(rotation=45)
plt.show()

# Create a bar plot of average salary by experience level
df.groupby('experience_level')['salary_in_usd'].mean().plot(kind='bar')
plt.title('Average Salary by Experience Level')
plt.xlabel('Experience Level')
plt.ylabel('Average Salary (USD)')
plt.xticks(rotation=45)
plt.show()

# Create a scatter plot of salary vs. years of experience
plt.scatter(df['experience_level'], df['salary_in_usd'])
plt.title('Salary vs. Years of Experience')
plt.xlabel('Years of Experience')
plt.ylabel('Salary (USD)')
plt.show()

# Create a box plot of salary by company location
df.boxplot('salary_in_usd', by='company_location')
plt.title('Salary by Company Location')
plt.xlabel('Company Location')
plt.ylabel('Salary (USD)')
plt.xticks(rotation=45)
plt.show()

# Create a histogram of salary
plt.hist(df['salary_in_usd'])
plt.title('Distribution of Salaries')
plt.xlabel('Salary (USD)')
plt.ylabel('Frequency')
plt.show()
```
